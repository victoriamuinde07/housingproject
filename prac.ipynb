```python
import pandas as pd
ames = pd.read_csv('ames.csv')
ames
```

```python
ames.isna().sum()
```

```python
ames["LotFrontage"].fillna(ames["LotFrontage"].mode()[0],inplace=True)
ames["LotFrontage"].isna().sum()
```

```python
ames.describe()
```

```python
ames_subset = ames[['LotArea','1stFlrSF','GrLivArea','SalePrice']]
print(ames_subset)
```

```python
ames_subset.columns
```

```python
ames_subset.corr()['SalePrice']
```

lotarea is slightly correlated to saleprice.if the lotarea increases the saleprice teds to increase slightly
the rest are highly positively correlated if they increase the saleprice increases too

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].scatter(ames["LotArea"], ames["SalePrice"])
axes[0].set_title('LotArea vs SalePrice')
axes[0].set_xlabel('LotArea')
axes[0].set_ylabel('SalePrice')

axes[1].scatter(ames["1stFlrSF"], ames["SalePrice"])
axes[1].set_title('1stFlrSF vs SalePrice')
axes[1].set_xlabel('1stFlrSF')
axes[1].set_ylabel('SalePrice')

# Plot for 'GrLivArea'
axes[2].scatter(ames["GrLivArea"], ames["SalePrice"])
axes[2].set_title('GrLivArea vs SalePrice')
axes[2].set_xlabel('GrLivArea')
axes[2].set_ylabel('SalePrice')

plt.tight_layout()
plt.show()
```

```python
import statsmodels.api as sm 
import statsmodels.graphics.regressionplots as sm_graphics
y = ames_subset ['SalePrice']
x = ames_subset[['LotArea','1stFlrSF','GrLivArea']]
x = sm.add_constant(x)
model = sm.OLS(y,x) 
results = model.fit() 
print(results.summary())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
X_sklearn = ames_subset[['LotArea', '1stFlrSF', 'GrLivArea']]

# Create and fit the LinearRegression model using scikit-learn
sklearn_ames_model = LinearRegression()
sklearn_ames_model.fit(X_sklearn, y)

# Calculate R-squared for scikit-learn model
r2_sklearn = sklearn_ames_model.score(X_sklearn, y)

# Print comparison results
print(f"""
StatsModels R-Squared:    {results.rsquared}
scikit-learn R-Squared:   {r2_sklearn}

StatsModels intercept:    {results.params["const"]}
scikit-learn intercept:   {sklearn_ames_model.intercept_}

StatsModels coefficient for LotArea:  {results.params["LotArea"]}
scikit-learn coefficient for LotArea: {sklearn_ames_model.coef_[0]}
""")
```
